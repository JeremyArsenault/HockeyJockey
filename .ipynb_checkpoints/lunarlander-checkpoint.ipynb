## Approach: Policy Gradients
Directly learn the policy by computing the derivative of the reward wrt/ the policy and do gradient ascent. Notice that we can't just use this model to estimate our future rewards anymore. The REINFORCE algorithm simulates the episode using the policy to observe the complete reward. We will be using the actor-critic algorithm, which trains a model to estimate the value of a current state (note that unlike the q learning value function, this value function learns the value of a state independent of actions,  which makes it much more easily learnable for large action spaces).

For DQN, we encouraged exploration / introduced stochasticity in training using an epsilon greedy strategy. This time we will train our model to output a mean for x/z thrust that we can sample from.

### Model:
Simple MLP Models for actor and critic

```python
class Actor(nn.Module):
    """
    Given state, finds action stats and selects action (gaussian)
    """    
    def __init__(self, state_dim, eps=1e-3):
        super(Actor, self).__init__()
        
        self.fc1 = nn.Linear(state_dim,32)
        self.fc1.weight.data.uniform_(-eps,eps)        
        self.fc2 = nn.Linear(32,64)
        self.fc2.weight.data.uniform_(-eps,eps)        
        self.fc3 = nn.Linear(64,16)
        self.fc3.weight.data.uniform_(-eps,eps)
        
        self.fcmx = nn.Linear(16,1)
        self.fcmx.weight.data.uniform_(-eps,eps)
        #self.fcvx = nn.Linear(16,1)
        #self.fcvx.weight.data.uniform_(-0.2,0.2)        
        self.fcmy = nn.Linear(16,1)
        self.fcmy.weight.data.uniform_(-eps,eps)       
        #self.fcvy = nn.Linear(16,1)
        #self.fcvy.weight.data.uniform_(-0.2,0.2)
        
    def action_stats(self, state):
        x = nn.functional.relu(self.fc1(state))
        x = nn.functional.relu(self.fc2(x))
        x = nn.functional.relu(self.fc3(x))
        
        mx = torch.tanh(self.fcmx(x))
        #vx = nn.functional.relu(self.fcvx(x))   
        vx = 0
        my = torch.tanh(self.fcmy(x))
        #vy = nn.functional.relu(self.fcvy(x))
        vy = 0
        return mx, vx, my, vy
        
    def forward(self, state):
        mx, vx, my, vy = self.action_stats(state)
        """
        px = torch.distributions.Normal(mx, vx)
        py = torch.distributions.Normal(my, vy)      
        a = torch.cat((px.rsample(), py.rsample()),dim=1)
        """
        a = torch.cat((mx,my),dim=1)
        return a
    
    def clip_grads(self, eps=1):
        for param in self.parameters():
            param.grad.data.clamp_(-eps, eps)
        """
        self.fc1.weight.grad.data.clamp_(-eps, eps)
        self.fc2.weight.grad.data.clamp_(-eps, eps)
        self.fc3.weight.grad.data.clamp_(-eps, eps)
        self.fcmx.weight.grad.data.clamp_(-eps, eps)
        #self.fcvx.weight.grad.data.clamp_(-eps, eps)
        self.fcmy.weight.grad.data.clamp_(-eps, eps)
        #self.fcvy.weight.grad.data.clamp_(-eps, eps)
        """
        
class Critic(nn.Module):
    """
    Q(s,a): Evaluates expect discounded sum of rewards of (s,a) pair
    """
    def __init__(self, state_dim, action_dim, eps=1e-3):
        super(Critic, self).__init__()
        
        self.fcs1 = nn.Linear(state_dim,64)
        self.fcs1.weight.data.uniform_(-eps,eps)
        self.fcs2 = nn.Linear(64,32)
        self.fcs2.weight.data.uniform_(-eps,eps)

        self.fca1 = nn.Linear(action_dim,32)
        self.fca1.weight.data.uniform_(-eps,eps)

        self.fc2 = nn.Linear(64,16)
        self.fc2.weight.data.uniform_(-eps,eps)

        self.fc3 = nn.Linear(16,1)
        self.fc3.weight.data.uniform_(-eps,eps)

    def forward(self, state, action):
        s1 = nn.functional.relu(self.fcs1(state))
        s2 = nn.functional.relu(self.fcs2(s1))
        a1 = nn.functional.relu(self.fca1(action))
        x = torch.cat((s2,a1),dim=1)

        x = nn.functional.relu(self.fc2(x))
        x = self.fc3(x)
        return x

    def clip_grads(self, eps=1):
        for param in self.parameters():
            param.grad.data.clamp_(-eps, eps)
        """
        self.fcs1.weight.grad.data.clamp_(-eps, eps)
        self.fcs2.weight.grad.data.clamp_(-eps, eps)
        self.fca1.weight.grad.data.clamp_(-eps, eps)
        self.fc2.weight.grad.data.clamp_(-eps, eps)
        self.fc3.weight.grad.data.clamp_(-eps, eps)
        """
```

```python
"""
helpful model.apply() functions
"""
def weights_init(m, std=1e-3):
    if isinstance(m, nn.Linear):
        torch.nn.init.normal_(m.weight, mean=0, std=std)
        torch.nn.init.zeros_(m.bias)
        
def clip_gradients(m):
    m.grad.data.clamp_(-1, 1)

def print_model(m):
    print(m)
    if isinstance(m, nn.Linear):
        print('Weight: ',m.weight, ' Bias: ',m.bias)
```

```python
def reward_transform(reward, gamma):
    """
    Because it's hard for our NN to learn to estimate these high reward
    values, it helps to normalize the rewards
    """
    return reward
    
def take_action(model, sd, state):
    """
    Return an action
    """
    model.eval()
    action_stats = model(torch.Tensor([state])).detach()[0]
    x = np.random.normal(action_stats[0], sd)
    z = np.random.normal(action_stats[1], sd)
    return np.array([x,z])

def uniform_sample(buffer, batch_size):
    """
    Sample uniformly from buffer
    """
    inds = np.random.choice(len(buffer), batch_size, replace=True)
    sample = [buffer[x] for x in inds]
    return sample, inds

def priority_sample(buffer, batch_size):
    """
    Sample highest TD priority batch from buffer
    """
    priorities = np.array([x[4] for x in buffer])
    inds = np.argsort(priorities)[-batch_size:]
    sample = [buffer[x] for x in inds]
    return sample, inds

def stoch_priority_sample(buffer, batch_size):
    """
    Sample from buffer using naive stochastic priority sampling
    """
    priorities = np.array([x[4] for x in buffer])*np.random.uniform(size=len(buffer))
    inds = np.argsort(priorities)[-batch_size:]
    sample = [buffer[x] for x in inds]
    return sample, inds

def sample_states(env, num_states=15, buff_size=60):
    """
    Sample random states from environment
    """
    buffer = []
    state = env.reset()
    while len(buffer)<buff_size:
        buffer.append(state)
        state, _ , done , _ = env.step(env.action_space.sample())
        if done:
            state = env.reset()
    sample, _ = uniform_sample(buffer, num_states)
    return sample

def compute_max_q(actor, critic, states):
    """
    Return the max model evaluation of sampled states
    """
    actor.eval()
    critic.eval()
    with torch.no_grad():
        s = torch.Tensor(states)
        q_vals = critic(s, actor(s))
    return float(q_vals.max())

def update_critic(actor, critic, buffer, batch_size, optimiser, discount_factor, exploration):
    """
    Do a single batch training update on actor
    """
    critic.eval()
    actor.eval()
    # sample buffer and preprocess batch
    # use critic to compute expected future sum of rewards
    batch, batch_inds = uniform_sample(buffer, batch_size)
    batch_states = torch.Tensor([x[0] for x in batch])
    batch_actions = torch.Tensor([x[1] for x in batch])
    batch_exp_rewards = []
    for b in batch:
        if b[2] is not None:
            with torch.no_grad():
                next_state = torch.Tensor([b[2]])
                next_action = actor(next_state)+torch.Tensor(np.random.normal(0,0.1*exploration,2))
                r = b[3] + discount_factor * critic(next_state, next_action)[0]
        else:
            r = b[3]
        batch_exp_rewards.append(reward_transform(r, discount_factor))
    batch_exp_rewards = torch.Tensor(batch_exp_rewards)
    
    # update model
    critic.train()  
    optimiser.zero_grad()
    td = torch.abs(critic(batch_states, batch_actions) - batch_exp_rewards)
    loss = torch.sum(torch.pow(td, 2)) #MSE
    #loss = torch.sum(td)  #MAE
    loss.backward()
    #critic.clip_grads()
    optimiser.step()
    
    # update buffer priorities
    td_sum = td.sum(1)
    for batch_i, buff_i in enumerate(batch_inds):
        b = list(buffer[buff_i])
        b[4] = float(td_sum[batch_i])
        buffer[buff_i] = tuple(b)

def update_actor(actor, critic, buffer, batch_size, optimiser, discount_factor):
    """
    Do a single batch training update on actor
    """
    # sample buffer and preprocess batch
    # use critic to compute expected future sum of rewards
    batch, batch_inds = uniform_sample(buffer, batch_size)
    batch_states = torch.Tensor([x[0] for x in batch])

    actor.train()
    critic.train()
    optimiser.zero_grad()
    a = actor(batch_states)
    loss = -1*torch.sum(critic(batch_states, a))
    loss.backward()
    #actor.clip_grads()
    optimiser.step()
    
def update(actor, critic, buffer, batch_size, actor_optimiser, critic_optimiser, discount_factor, exploration):
    critic.eval()
    actor.eval()
    # sample buffer and preprocess batch
    # use critic to compute expected future sum of rewards
    batch, batch_inds = uniform_sample(buffer, batch_size)
    batch_states = torch.Tensor([x[0] for x in batch])
    batch_actions = torch.Tensor([x[1] for x in batch])
    batch_exp_rewards = []
    for b in batch:
        if b[2] is not None:
            with torch.no_grad():
                next_state = torch.Tensor([b[2]])
                next_action = actor(next_state)
                r = b[3] + discount_factor * critic(next_state, next_action)[0]
        else:
            r = b[3]
        batch_exp_rewards.append(reward_transform(r, discount_factor))
    batch_exp_rewards = torch.Tensor(batch_exp_rewards)
    
    # update model
    critic.train()  
    critic_optimiser.zero_grad()
    td = torch.abs(critic(batch_states, batch_actions) - batch_exp_rewards)
    loss = torch.sum(torch.pow(td, 2)) #MSE
    #loss = torch.sum(td)  #MAE
    loss.backward()
    critic.clip_grads()
    critic_optimiser.step()
    
    with torch.no_grad():
        means = actor(batch_states)
    actor.train()
    critic.train()
    actor_optimiser.zero_grad()
    a = actor(batch_states)
    loss = -1*torch.sum(critic(batch_states, a))
    #e =  torch.sum((batch_actions-means)/pow(0.05,2) * actor(batch_states),1)
    #e =  torch.sum((batch_actions-actor(batch_states))/pow(exploration,2),1)
    #loss = -1*torch.sum(e * critic(batch_states, batch_actions))
    loss.backward()
    actor.clip_grads()
    actor_optimiser.step()


def pg_train(env, actor, critic, actor_optimiser, critic_optimiser, episodes, buffer_size=500, batch_size=32, discount_factor=0.9, verbose=True):
    """
    Simulate and train model to maximize reward from env
    """
    q_states = sample_states(env)
    q_history = []
    loss_history = []
    reward_history = []
    initial_priority = 10

    buffer = []
    for episode in range(episodes):
        i = 0
        done = False
        current_state = env.reset()
        while not done:
            # take step and append to buffer
            # if step is terminal set next state to None, reward to 0
            # set priority to high since we haven't trained on this point yet
            exploration=0.2*(1-episode/episodes)
            with torch.no_grad():
                action = actor(torch.Tensor([current_state]))[0].numpy()+np.random.normal(0,exploration,2)
            #action = take_action(actor, actor_sd, current_state)
            next_state, reward, done, _  = env.step(action)
            if done:
                next_state = None
            
            buffer.append((current_state, action, next_state, reward/25, initial_priority))
            current_state = next_state

            # if buffer isn't fully populated keep simulating
            # otherwise make sure the right size
            i += reward                
            if len(buffer)>buffer_size:
                buffer.pop(0)
            elif len(buffer)<buffer_size//2:
                continue
              
            # do a training step
            # update_critic(actor, critic, buffer, batch_size, critic_optimiser, discount_factor, exploration=(1-episode/episodes))
            # update_actor(actor, critic, buffer, batch_size, actor_optimiser, discount_factor)
            update(actor, critic, buffer, batch_size, actor_optimiser, critic_optimiser, discount_factor, exploration)
            
        # update performance metrics
        reward_history.append(i)
        q = compute_max_q(actor, critic, q_states)
        q_history.append(q)
            
        if verbose==True and episode%100==0:
            print('Episode: ',episode,'  reward: ',i)
            
    return q_history, reward_history
```

```python
actor_lr=1e-3
critic_lr=1e-3
weight_init=1e-3

env = gym.make('LunarLanderContinuous-v2')

actor = Actor(8, eps=weight_init)
critic = Critic(8,2, eps=weight_init)

actor_optim = torch.optim.SGD(actor.parameters(), lr=actor_lr)
critic_optim = torch.optim.SGD(critic.parameters(), lr=critic_lr)

q_history, reward_history = pg_train(env, actor, critic, actor_optim, critic_optim, 3000, buffer_size=1500, batch_size=64, discount_factor=0.99)

plt.plot(q_history, label='rewards')
plt.title('Q')
plt.show()
```
